## Representación gráfica de datos

Si a la función de trazado se le da una matriz de datos, la usará como coordenadas en el eje vertical, y utilizará el índice de cada punto de datos en el array como la coordenada horizontal.

```python
lista =[1,2,5,7,8,3,1]
#los representa en el eje y
plt.plot(lista)
plt.show()
```

También se puede proporcionar dos matrices: una para el eje horizontal, y otra para el eje vertical. (ambas listas deben tener la misma dimencion)


```python
lista2=[1,4,6,7,8,9,10]
plt.plot(lista2,lista)
plt.show()
```

Pueden modificarse las logitudes de los ejes para que la figura no se vea tan ajustada

```python
plt.plot([-3,-1,0,4,7],[1,4,6,7,8])
plt.show()
```

```python
plt.plot([-3,-1,0,4,7],[1,4,6,7,8])
plt.axis([-4,8,0,10])
#minimo eje x, maximo eje x y,minimo eje y, maximo eje y
```

Se sigue el mismo procedimiento para pintar una función matemática

```python
import numpy as np
#de -2 a 2 con 500 elementos
x=np.linspace(-2,2,500)
x
```

También pude modificarse el estilo de la gráfica para que contenga más información.

```python
y=x**2
plt.plot(x,y)
plt.show()
```

Pueden superponerse gráficas y cambiar el estilo de las líneas

```python
plt.plot(x,y)
plt.title("Squeare Function")
plt.xlabel("x")
plt.ylabel("y")
plt.grid(True)
plt.show()
```

```python
# Separando en diferentes lineas las funciones
x= np.linspace(-2,2,500)
y=x**2
y2=x+1
plt.plot(x,y,'b--',x,y2,'g p')
plt.show()
```

Para poder diferenciar entre ambas funciones siempre es recomendable añadir una leyenda

```python
x= np.linspace(-2,2,500)
y=x**2
y2=x+1
plt.plot(x,y,'b--')
plt.plot(x,y2,'g')
plt.show()
```

```python
x= np.linspace(-2,2,500)
y=x**2
y2=x+1
plt.plot(x,y,'b--',label="x^2")
plt.plot(x,y2,'g',label="x+1")
plt.legend(loc="best")#cual es el mejor lugra para mostrar las leyendas
plt.show()
```

Tambien puede crearse dos graficas que no se superpongan. Estas graficas se organizan en un grid y se denominan subplots.

```python
x=np.linspace(-2,2,500)
y=x**2
y2=x+1

plt.subplot(1,2,1)
plt.plot(x,y,'b--')

plt.subplot(1,2,2)
plt.plot(x,y2,'g')

plt.show()
```

```python
plt.figure(figsize=(14,6))#ancho x alto

plt.subplot(1,2,1)
plt.plot(x,y,'b--')
plt.xlabel("x1" ,fontsize=14)
plt.ylabel("y1",fontsize=14)

plt.subplot(1,2,2)
plt.plot(x,y2,'g')
plt.xlabel("x")
plt.ylabel("y2")


plt.show()
```

Para que las gráficas no queden tan ajustadas, podemos hacer la figura más grande.

## Histogramas

```python
#grafico de barras
data = [1, 1.1, 1.8, 2, 2.1, 3.2, 3, 3, 3, 3]
#fila, columna , subplot
plt.subplot(211)
#lista,bins son mas barras,ancho
plt.hist(data,bins=30,rwidth=0.8)
plt.xlabel('VALOR')
plt.ylabel('FRECUENCIA')
plt.show()

```

```python

data = [1, 1.1, 1.8, 2, 2.1, 3.2, 3, 3, 3, 3]
#fila, columna , subplot
plt.subplot(211)
#lista,bins son mas barras,ancho
plt.hist(data,bins=30,rwidth=0.8)
plt.xlabel('VALOR')
plt.ylabel('FRECUENCIA')
plt.show()

plt.subplot(212)
plt.hist(data,bins=[1.5,2,2.5,3],rwidth=0.95)

#lista,bins son mas barras,ancho

plt.xlabel('VALOR 1')
plt.ylabel('FRECUENCIA 2')
plt.show()
```
