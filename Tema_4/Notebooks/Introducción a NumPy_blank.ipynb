## Arrays

Un **array** es una estructura de datos que consiste en una colección de elementos (valores o variables), cada uno identificado por al menos un índice o clave. Un array se almacena de modo que la posición de cada elemento se pueda calcular a partir de su tupla de índice mediante una fórmula matemática. El tipo más simple de array es un array lineal, también llamado array unidimensional.

En numpy:
* Cada dimensión se denomina **axis**
* El número de dimensiones se denomina **rank**
* La lista de dimensiones con su correspondiente longitud se denomina **shape**
* El número total de elementos (multiplicación de la longitud de las dimensiones) se denomina **size**

```python
# Array cuyos valores son todos 0
a=np.zeros((2,4))
```

```python
a
```

_**a**_ es un array:
* Con dos **axis**, el primero de longitud 2 y el segundo de longitud 4
* Con un **rank** igual a 2
* Con un **shape** igual (2, 4)
* Con un **size** igual a 8

## Acceso a los elementos de un array

### Array unidimensional

```python
# Creación de un Array unidimensional
array_uni=np.array([1,3,5,7,9,11])

print("Shape:",array_uni.shape ) #corresponde a la cantidad de elementos del array
print("Array_uni:", array_uni)#imprime el array unidimensional creado
```

```python
# Accediendo al quinto elemento del Array
array_uni[4]
```

```python
# Accediendo al tercer y cuarto elemento del Array
array_uni[2:4]#slice que accede a los elementos desde el índice 2 hasta el índice 3 (excluyendo el índice 4), es decir, accede a los elementos en las posiciones 2 y 3 del array.
```

```python
# Accediendo a los elementos 0, 3 y 5 del Array
array_uni[0:6:2]#stride de 2, accede a los elementos desde el índice 0 hasta el índice 5 (excluyendo el índice 6) con un paso de 2, es decir, accede a los elementos en las posiciones 0, 2 y 4 del array.
```

### Array multidimensional

```python
# Creación de un Array multidimensional
array_multi=np.array([[1,2,3,4],[5,6,7,8]])
print("Shape:",array_multi.shape) #corresponde a la cantidad de filas y columnas del array
print("Array_multi:", array_multi)#imprime el array multidimensional creado
```

```python
# Accediendo al cuarto elemento del Array
array_multi[0,3]#accede al elemento en la fila 0 y columna 3 del array multidimensional, es decir, accede al número 4.
```

```python
# Accediendo a la segunda fila del Array
array_multi[1,:]#accede a la fila 1 (segunda fila) del array multidimensional, y el símbolo ":" indica que se seleccionan todos los elementos de esa fila, es decir, accede a los números 5, 6, 7 y 8.
```

```python
# Accediendo al tercer elemento de las dos primeras filas del Array
array_multi[0:2,2]#accede a los elementos en la columna 2 (tercera columna) de las filas 0 y 1 (primeras dos filas)
#del array multidimensional, es decir, accede a los números 3 y 7.
```

```python
array_multi[0:,2]#accede a los elementos en la columna 2 (tercera columna) de todas las filas del array multidimensional, es decir, accede a los números 3 y 7.
```

## Modificación de un Array

```python
# Creación de un Array unidimensional inicializado con el rango de elementos 0-27
array1=np.arange(28)
print("Shape:",array1.shape) #corresponde a la cantidad de elementos del array
print("Array1:", array1)#imprime el array unidimensional creado
```

```python
# Cambiar las dimensiones del Array y sus longitudes
array1.shape =(7,4) 
print("Shape:",array1.shape) #corresponde a la cantidad de filas y columnas del array
print("Array1:", array1)#imprime el array unidimensional creado

#cambia la forma del array1 a una matriz de 7 filas y 4 columnas, reorganizando los elementos del array original en esta nueva estructura.
```

```python
# El ejemplo anterior devuelve un nuevo Array que apunta a los mismos datos. 
# Importante: Modificaciones en un Array, modificaran el otro Array
array2=array1.reshape(4,7)
print("Shape:",array2.shape)
print("Array2:\n", array2)

```

```python
# Modificación del nuevo Array devuelto
array2[0,3] = 20
print("Array2:\n", array2)
```

```python
print("Array1:\n", array1)
```

```python
# Desenvuelve el Array, devolviendo un nuevo Array de una sola dimension
# Importante: El nuevo array apunta a los mismos datos
print("Array1:\n", array1.ravel())
```

## Operaciones aritméticas con Arrays

```python
# Creación de dos Arrays unidimensionales
array1=np.arange(2,18,2)#crea un array unidimensional con valores que comienzan en 2, terminan antes de 18 y se incrementan en pasos de 2, 
#es decir, crea un array con los números pares del 2 al 16.
array2=np.arange(8)

print("Array1:",array1)
print("Array2:",array2)
```

```python
# Suma
print("Suma:", array1 + array2) #suma los elementos correspondientes de array1 y array2, es decir, 
#suma el primer elemento de array1 con el primer elemento de array2, el segundo elemento de array1 con 
# el segundo elemento de array2, y así sucesivamente, devolviendo un nuevo array que contiene los resultados de estas sumas. 
# a diferencia de las listas 
```

```python
# Resta
print("Resta:", array1 - array2) #resta los elementos correspondientes de array1 y array2, es decir,
#resta el primer elemento de array1 con el primer elemento de array2, el segundo elemento de array1 con el segundo elemento de array2, 
# y así sucesivamente, devolviendo un
```

```python
# Multiplicacion
# Importante: No es una multiplicación de matrices
print("Multiplicacion:", array1 * array2) #multiplica los elementos correspondientes de array1 y array2, es decir,
#multiplica el primer elemento de array1 con el primer elemento de array2, el segundo elemento de array1 con el segundo elemento de array2, y así sucesivamente, devolviendo un
```

## Broadcasting

Si se aplican operaciones aritméticas sobre Arrays que no tienen la misma forma (shape) Numpy aplica un propiedad que se denomina Broadcasting.

```python
# Creación de dos Arrays unidimensionales
array1=np.arange(5)
array2=np.array([3])
print("ShapeArray1:", array1.shape)
print("Array1:", array1)
print("ShapeArray2:", array2.shape)
print("Array2:", array2)
print()
```

```python
# Suma de ambos Arrays
array1 + array2 #suma cada elemento de array1 con el único elemento de array2 (que es 3), devolviendo un nuevo array donde cada elemento es el resultado de esta suma.     
```

```python
# Creación de dos Arrays multidimensional y unidimensional
array1=np.arange(6)
array1.shape=(6,18,4)
print("ShapeArray1:", array1.shape)
print("Array1:\n", array1)
print()
print("ShapeArray2:", array2.shape)
print("Array2:", array2)
print()
```

```python
# Suma de ambos Arrays
array1 + array2 #suma cada elemento de array1 con el único elemento de array2 (que es 3), devolviendo un nuevo array donde cada elemento es el resultado de esta suma.
```

## Funciones estadísticas sobre Arrays

```python
# Creación de un Array unidimensional
array1=np.arange(1,20,2)
print("Array:", array1  )
```

```python
# Media de los elementos del Array
array1.mean() #calcula la media (promedio) de los elementos del array1, sumando todos los elementos y dividiendo por la cantidad total de elementos, devolviendo el resultado como un valor escalar.
```

```python
# Suma de los elementos del Array
array1.sum() #calcula la suma de todos los elementos del array1, devolviendo el resultado como un valor escalar.
```

Funciones universales eficientes proporcionadas por numpy: **ufunc**

```python
# Cuadrado de los elementos del Array
array1.square(array1) #calcula el cuadrado de cada elemento del array1, devolviendo un nuevo array donde cada elemento es el resultado de elevar al cuadrado el elemento correspondiente del array original.
```

```python
# Raiz cuadrada de los elementos del Array
np.sqrt(array1) #calcula la raíz cuadrada de cada elemento del array1, devolviendo un nuevo array donde cada elemento es el resultado de esta operación.
```

```python
# Exponencial de los elementos del Array
np.exp(array1) #calcula la función exponencial de cada elemento del array1, devolviendo un nuevo array donde cada elemento es el resultado de esta operación.
```

```python
# log de los elementos del Array
np.log(array1) #calcula el logaritmo natural de cada elemento del array1, devolviendo un nuevo array donde cada elemento es el resultado de esta operación.
```
